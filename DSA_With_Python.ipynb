### **Creating our own list**

```python
import ctypes
```

```python
class MeraList:

  def __init__(self):
    self.size = 1
    self.n = 0
    # create a C type array with size = self.size
    self.A = self.__make_array(self.size)

  # Adding len functionality to our list class
  def __len__(self):
    return self.n

  # Adding print functionality
  def __str__(self):
    result = ''
    for i in range(self.n):
      result = result + str(self.A[i]) + ','
    if self.n > 0:
      return '[' + result[:-1] + ']'
    else:
      return '[]'

  # Fetch item using index
  def __getitem__(self,index):
    if 0 <= index < self.n:
      return self.A[index]
    else:
      return 'IndexError - Index out of range'

  # Deleting item form an array (renamed and fixed)
  def __delitem__(self,index):
    if not 0 <= index < self.n:
      return 'del: index out of range'
    # Shift elements to the left
    for i in range(index,self.n - 1):
      self.A[i] = self.A[i + 1]
    self.n = self.n - 1

  # Adding append function
  def append(self,item):
    if self.n == self.size:
      # resize
      self.__resize(self.size*2)
    # append
    self.A[self.n] = item
    self.n = self.n + 1

  # Adding pop function
  def pop(self):
    if self.n == 0:
      return 'Empty list'
    popped_item = self.A[self.n - 1]
    self.n = self.n - 1
    return popped_item

  # Adding clear function
  def clear(self):
    self.n = 0
    self.size = 1
    self.A = self.__make_array(self.size)

  # Searching an item in an array
  def find(self,item):
    for i in range(self.n):
      if self.A[i] == item:
        return i
    return 'ValueError - Not in list'

  # Inserting item in an array - middle
  def insert(self,pos,item):
    if 0 <= pos < self.n:
      # Resize if necessary
      if self.n == self.size:
        self.__resize(self.size*2)

      for i in range(self.n - 1, pos - 1, -1):
        self.A[i + 1] = self.A[i]

      self.A[pos] = item
      self.n = self.n + 1
    else:
      return 'IndexError'

  # Removing Item by value
  def remove(self,item):
    pos = self.find(item)

    if type(pos) == int:
      # delete
      self.__delitem__(pos)
    else:
      return pos


  # Resize Function
  def __resize(self,new_capacity):
    # create a new array with new capacity
    B = self.__make_array(new_capacity)
    self.size = new_capacity
    # copy the content of A to B
    for i in range(self.n):
      B[i] = self.A[i]
      # reassign A
    self.A = B

  def __make_array(self,capacity):
    # create a C array(static, referential) with size capacity
    return (capacity*ctypes.py_object)()
```
